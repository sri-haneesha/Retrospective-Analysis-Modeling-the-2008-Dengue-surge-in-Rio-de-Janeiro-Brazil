```python
#Temperatures from here: https://www.wunderground.com/history/weekly/br/rio-de-janeiro/SBRJ/date/2007-9-1

#temp starting from sept 1, 2007 in F
# temp = [73,70.5, 71.8, 75, 74.3, 74.5, 72.5, 73, 72.4, 73.3, 72.4, 73.1, 73.5, 73, 74.2, 70.8, 70.4, 69.6, 70.3, 72, 72.4, 72.8, 75.3, 74.7, 66.2, 65.5, 70.5, 71.7, 69.3, 68 ]

#2007 data
sept = [73.2, 70.6, 71.8, 75.0, 74.3, 74.5, 72.5, 73.0, 72.4, 73.3,
    72.4, 73.1, 73.5, 73.0, 74.2, 70.6, 70.4, 69.6, 70.3, 72.0,
    72.4, 72.8, 75.2, 74.7, 66.2, 65.5, 70.5, 71.7, 69.3, 68.1, 68.0]

print(len(sept))

oct = [
    66.0, 68.9, 68.8, 70.2, 70.9, 72.9, 71.9, 69.8, 72.4, 74.6,
    70.3, 71.5, 73.2, 73.4, 71.3, 74.8, 76.2, 70.9, 68.9, 71.4,
    75.9, 75.7, 71.8, 70.2, 70.0, 73.3, 74.9, 75.7, 77.0, 75.3, 76.5
]

print(len(oct))

nov = [
    80.2, 78.2, 79.3, 75.0, 70.0, 71.6, 74.7, 74.1, 78.7, 80.9,
    78.8, 70.3, 68.6, 76.3, 72.5, 67.6, 70.3, 73.8, 75.5, 76.9,
    73.4, 70.0, 70.6, 74.4, 78.4, 75.1, 72.9, 71.1, 70.8, 72.2
]

print(len(nov))

dec = [
    72.7, 74.2, 74.9, 76.2, 77.9, 80.8, 74.3, 77.8, 79.3, 80.4,
    83.6, 78.6, 71.7, 73.2, 76.9, 77.0, 77.5, 77.4, 77.3, 74.8,
    75.6, 76.2, 74.5, 77.0, 80.1, 81.8, 83.0, 81.8, 81.4, 83.4, 82.3
]

print(len(dec))

#2008 data

jan = [79.5, 80.2, 82.8, 84.5, 79.5, 75.8, 75.2, 77.3, 77.4, 77.3,
81.3, 82.9, 78.8, 79.3, 76.5, 80.2, 81.8, 80.3, 82.2, 78.1,
69.2, 69.0, 69.5, 74.8, 75.1, 72.0, 74.0, 71.4, 70.2, 72.6, 76.4]

print(len(jan))

feb = [77.1, 79.2, 74.6, 72.5, 74.3, 72.2, 75.6, 75.5, 76.6, 79.0,
 80.4, 80.9, 80.0, 78.6, 78.8, 79.2, 79.6, 77.3, 76.6, 77.0,
 79.3, 78.3, 77.7, 76.0, 76.9, 74.4, 77.2, 76.1, 76.5]

print(len(feb))

mar = [75.5, 81.3, 81.6, 81.4, 79.2, 78.6, 79.6, 79.9, 78.5, 77.7,
79.9, 78.0, 77.6, 70.7, 68.8, 72.0, 75.6, 77.4, 77.9, 77.9,
78.3, 79.9, 77.9, 76.7, 74.9, 74.4, 75.3, 75.5, 78.3, 75.7, 74.1, 70.0]

print(len(mar))



# Concatenating the lists of average temperatures in Fahrenheit for September to December
temps_fahrenheit = sept + oct + nov + dec + jan + feb + mar

# Converting the temperatures from Fahrenheit to Celsius
temps_celsius = [(temp - 32) * 5.0/9.0 for temp in temps_fahrenheit]

# temps_celsius = np.random.uniform(21, 28, 365*6)
print(len(temps_celsius))
print(temps_celsius)

length_of_temps_array = len(temps_celsius)


# Create a simulated precipitation data array with random values between 0.2 and 1.4
precipitation_data = np.random.uniform(0.2, 1.4, length_of_temps_array)

# Convert to a list
precipitation_data_list = precipitation_data.tolist()

print(precipitation_data_list)
```

```python
#Initial Conditions

NH = 6093472  # Total human population
NV = 3 * NH


E_s0 = 1000
L_s0 = 1000
P_s0 = 1000

E_i0 = 0
L_i0 = 0
P_i0 = 0

A_em_s0 = NV/4
A_h_s0 = NV/4
A_g_s0 = NV/4
A_o_s0 = NV/4

A_h_e0 = 0
A_g_e0 = 0
A_o_e0 = 0

A_em_i0 = 0
A_h_i0 = 0
A_g_i0 = 0
A_o_i0 = 0


H_s0 = (NH - 0) / NH  # All except 0 are susceptible
H_e0 = 0 / NH         # None exposed initially
H_i0 = 0 / NH        # 0 infected initially
H_r0 = 0 / NH         # None recovered initially
```

```python
#Funtions for params that are temp dependent
def biting_rate(T, b0 = 0.001, b1 = 0.001):
    return b0 + b1 * T

def f_E_fn(T, T_E, TDD_E):
    if T > T_E:
        return (T - T_E) / TDD_E
    else:
        return 0

def f_Ag_fn(T, T_Ag, TDD_Ag):
    if T > T_Ag:
        return (T - T_Ag) / TDD_Ag
    else:
        return 0

def f_P_fn(T):
    return 0.021 * (np.exp(0.162*(T - 10)) - np.exp(0.162*(35 - 10) - (35 - T)/5.007))

def f_L_fn(T):
    return f_P_fn(T) / 4

def k_L_fn(R, alpha_L0, alpha_L1):
    return alpha_L0 + alpha_L1 * R

def k_P_fn(R, alpha_P0, alpha_P1):
    return alpha_P0 + alpha_P1 * R

def m_L_fn(T, mu_L):
    return np.exp(-T / 2) + mu_L

def m_P_fn(T, mu_P):
    return np.exp(-T / 2) + mu_P

# mA(T) function with conditions
def m_A_fn(T, mu_A):
    m_A_T = 0.1 - 0.00667 * T + 0.0001487 * T**2
    return max(m_A_T, mu_A)


#Params


beta = 300
rho = 0.078
sigma = 0.5
d_tp = 0.216
d_mi = 0.1
r = 1/5
mu_E = 0.1
mu_L = 0.08
mu_P = 0.1
mu_em = 0.17
mu_A = 1/30


T_E = 16
T_DD_E = 42
gamma_Aem = 0.25
gamma_Ao = 2
gamma_H = 1/5.5
T_Ag = 9.9
T_DD_Ag = 36


immunity = 1/60

```

```python
t0 = 30
CR = 0.40
```

```python
# Differential equations in Python

def dengue_model(y, t, params):

    # Get the temperature for the current day
    day_index = int(t) % len(temps_celsius)  # Loop back after one year
    T = temps_celsius[day_index]
    R = precipitation_data_list[day_index]


    # Unpack the current values of the variables
    E_s, L_s, P_s, E_i, L_i, P_i, A_em_s, A_h_s, A_g_s, A_o_s, A_h_e, A_g_e, A_o_e, A_em_i, A_h_i, A_g_i, A_o_i, H_s, H_e, H_i, H_r = y

    # Unpack parameters
    beta, rho, sigma, mu_E, mu_L, mu_P, mu_em, mu_A, gamma_Aem, gamma_Ao, d_tp, d_mi, r = params

    #Calculating temp dependent functions

    b = biting_rate(T, 0.001, 0.001)
    f_E = f_E_fn(T, T_E, T_DD_E)
    f_L = f_L_fn(T)
    f_P = f_P_fn(T)
    f_Ag = f_Ag_fn(T, T_Ag, T_DD_Ag)
    k_L = k_L_fn(R, 10 ** 12, 10 ** 12)
    k_P = k_P_fn(R, 10 ** 12, 10 ** 12)
    I_p = 0.216 #Find about it in page 2 right side: d_tp value
    m_L = m_L_fn(T, mu_L)
    m_P = m_P_fn(T, mu_P)
    m_A = m_A_fn(T, mu_A)

    # Derivatives
    dE_s_dt = beta * (A_o_s + A_o_e + (1 - rho) * A_o_i) - (mu_E + f_E) * E_s
    dL_s_dt = f_E * E_s - (m_L * (1 + (L_s + L_i) / k_L) + f_L) * L_s
    dP_s_dt = f_L * L_s - (m_P + f_P) * P_s

    dE_i_dt = beta * rho * A_o_i - (mu_E + f_E) * E_i if t >= t0 else 0
    dL_i_dt = f_E * E_i - (m_L * (1 + (L_s + L_i) / k_L) + f_L) * L_i if t >= t0 else 0
    dP_i_dt = f_L * L_i - (m_P + f_P) * P_i if t >= t0 else 0

    dA_em_s_dt = f_P * P_s * sigma * np.exp(-mu_em * (1 + (P_s + P_i) / k_P)) - (m_A + gamma_Aem) * A_em_s # sus exp?
    dA_h_s_dt = gamma_Aem * A_em_s + gamma_Ao * A_o_s - (m_A + b) * A_h_s
    dA_g_s_dt = b * (I_p * (1 - d_tp) + (1 - I_p)) * A_h_s - (m_A + f_Ag) * A_g_s
    dA_o_s_dt = f_Ag * A_g_s - (m_A + gamma_Ao) * A_o_s

    dA_h_e_dt = gamma_Ao * (1 - d_mi) * A_o_e - (m_A + b) * A_h_e if t >= t0 else 0
    dA_g_e_dt = b * A_h_e * (1 - d_mi) + b * d_tp * I_p * A_h_s - (m_A + f_Ag) * A_g_e if t >= t0 else 0
    dA_o_e_dt = f_Ag * (1 - d_mi) * A_g_e - (m_A + gamma_Ao) * A_o_e if t >= t0 else 0

    dA_em_i_dt = f_P * P_i * sigma * np.exp(-mu_em * (1 + (P_s + P_i) / k_P)) - (m_A + gamma_Aem) * A_em_i #here if t >= t0 else 0
    dA_h_i_dt = gamma_Aem * A_em_i + gamma_Ao * (A_o_i + A_o_e) + b * d_mi * A_h_e - (m_A + b) * A_h_i if t >= t0 else 0
    # dA_h_i_dt = (gamma_Aem * A_em_i + gamma_Ao * (A_o_i + A_o_e) + b * d_mi * A_h_e - (m_A + b) * A_h_i + 50) if t == t0 else (gamma_Aem * A_em_i + gamma_Ao * (A_o_i + A_o_e) + b * d_mi * A_h_e - (m_A + b) * A_h_i if t > t0 else 0)
    dA_g_i_dt = b * (A_h_i + d_mi * A_h_e) - (m_A + f_Ag) * A_g_i if t >= t0 else 0
    dA_o_i_dt = f_Ag * (A_g_i + d_mi * A_g_e) - (m_A + gamma_Ao) * A_o_i if t >= t0 else 0

    # dH_s_dt = -b * d_tp * A_h_i * H_s + sigma * H_r
    # dH_e_dt = b * d_tp * A_h_i * H_s - gamma_H * H_e if t >= t0 else 0
    # dH_i_dt = gamma_H * H_e - r * H_i if t >= t0 else 0
    # dH_r_dt = r * H_i - sigma * H_r if t >= t0 else 0



    dH_s_dt = -b * d_tp * CR * A_h_i * H_s + immunity * H_r
    dH_e_dt = b * d_tp * CR * A_h_i * H_s - gamma_H * H_e if t >= t0 else 0
    # dH_i_dt = gamma_H * H_e - r * H_i if t >= t0 else 0
    dH_i_dt = gamma_H * H_e - r * H_i + (7 if t == t0 else 0) if t >= t0 else 0
    dH_r_dt = r * H_i - immunity * H_r if t >= t0 else 0



    return [dE_s_dt, dL_s_dt, dP_s_dt,
            dE_i_dt, dL_i_dt, dP_i_dt,
            dA_em_s_dt, dA_h_s_dt, dA_g_s_dt, dA_o_s_dt,
            dA_h_e_dt, dA_g_e_dt, dA_o_e_dt,
            dA_em_i_dt, dA_h_i_dt, dA_g_i_dt, dA_o_i_dt,
            dH_s_dt, dH_e_dt, dH_i_dt, dH_r_dt]

# Initial conditions
initial_conditions = [
    E_s0,  # Initial susceptible eggs
    L_s0,  # Initial susceptible larvae
    P_s0,  # Initial susceptible pupae
    E_i0,  # Initial infected eggs
    L_i0,  # Initial infected larvae
    P_i0,  # Initial infected pupae
    A_em_s0,  # Initial susceptible emerging adults
    A_h_s0,  # Initial susceptible host-seeking adults
    A_g_s0,  # Initial susceptible gonotrophic adults
    A_o_s0,  # Initial susceptible ovipositing adults
    A_h_e0,  # Initial exposed host-seeking adults
    A_g_e0,   # Initial exposed gonotrophic adults
    A_o_e0,  # Initial exposed ovipositing adults
    A_em_i0,  # Initial infected emerging adults
    A_h_i0,  # Initial infected host-seeking adults
    A_g_i0, # Initial infected gonotrophic adults
    A_o_i0, # Initial infected ovipositing adults
    H_s0,  # Initial susceptible humans
    H_e0,  # Initial exposed humans
    H_i0,  # Initial infected humans
    H_r0   # Initial recovered humans
]

# Parameters
params = [
    beta,  # Number of eggs laid by ovipositing females
    rho,   # Vertical transmission probability
    sigma,  # Sex-ratio at emergence
    mu_E,  # Egg mortality rate
    mu_L,  # Larva mortality rate
    mu_P,  # Pupa mortality rate
    mu_em,  # Emerging adult mortality rate
    mu_A,  # Adult mortality rate
    gamma_Aem,  # Emerging adult development rate
    gamma_Ao,  # Transition rate from ovipositing site-seeking to host-seeking adults
    d_tp,  # Dengue transmission probability per bite
    d_mi,  # Dengue incubation rate in mosquitoes
    r,  # Recovery rate from dengue, Humans

]


# Time vector
t = np.linspace(0, len(temps_celsius) - 1, len(temps_celsius))

# Solve ODE
solution = odeint(dengue_model, initial_conditions, t, args=(params,))


# Extract solutions
E_s, L_s, P_s, E_i, L_i, P_i, A_em_s, A_h_s, A_g_s, A_o_s, A_h_e, A_g_e, A_o_e, A_em_i, A_h_i, A_g_i, A_o_i, H_s_percentage, H_e_percentage, H_i_percentage, H_r_percentage = solution.T

# Convert percentages to actual numbers for human compartments
H_s = H_s_percentage * NH
H_e = H_e_percentage * NH
H_i = H_i_percentage * NH
H_r = H_r_percentage * NH

print(len(H_s))
```
